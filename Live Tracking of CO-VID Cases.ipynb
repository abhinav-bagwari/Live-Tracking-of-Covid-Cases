```python
import pandas as pd
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
df
```

```python
df=df.iloc[:-1,1:]
```

```python
df.tail()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
total_states = np.arange(len(df['state_name']))
```

```python
total_states
```

```python
df.active
```

#### Total Positve Cases 

```python
from matplotlib.pyplot import figure
figure(num=None,figsize=(8,10), facecolor='w', edgecolor = 'k')
plt.barh(total_states,df.positive, align='center', alpha=0.5,color=(1,0,0), edgecolor=(0.5,0.2,0.8))

plt.yticks(total_states, df['state_name'])
plt.xlim(1,max(df.positive) +10)
plt.xlabel('Positive Number of Cases')
plt.title('Corona Virus Cases')
plt.show()
```

#### Total Active Number of Cases

```python
from matplotlib.pyplot import figure
figure(num=None,figsize=(8,10), facecolor='w', edgecolor = 'k')
plt.barh(total_states,df.active, align='center', alpha=0.5,color=(1,0,0), edgecolor=(0.5,0.2,0.8))

plt.yticks(total_states, df['state_name'])
plt.xlim(1,max(df.active) +10)
plt.xlabel('Active Number of Cases')
plt.title('Corona Virus Cases')
plt.show()
```

#### Total Death Number of Cases

```python
from matplotlib.pyplot import figure
figure(num=None,figsize=(8,10), facecolor='w', edgecolor = 'k')
plt.barh(total_states,df.death, align='center', alpha=0.5,color=(0,0,1), edgecolor=(0.5,0.2,0.8))

plt.yticks(total_states, df['state_name'])
plt.xlim(1,max(df.death) +10)
plt.xlabel('Active Number of Cases')
plt.title('Corona Virus Cases')
plt.show()
```
